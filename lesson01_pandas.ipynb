```python
# Sắp xếp dữ liệu
newdf.sort_values(by='T5', ascending=True)
```

```python
# Lấy dữ liệu theo điều kiện
newdf[newdf['T6'] >= 8]
newdf[(newdf['T6'] > 7) & (newdf['T6'] < 9)]
newdf.loc[newdf['KT'] == 'C']
```

```python
# Part: Data Aggregation
# Nguyên lý ứng dụng các hàm tổng hợp trên các biến định tính và định lượng
# Định tính: count, mode
# Định lượng: count, mode, sum, mean
newdf.aggregate({'T5':['min', 'max'], 'T6': ['mean', 'sum']})
```

```python
# Gom nhóm: Nguyên tắc gom nhóm
newdf.groupby(['KV'])['KV'].agg(['count'])
```

```python
newdf.groupby(['KV'])['T5'].agg(['count', 'min', 'mean', 'std', 'max', 'sum'])
```

```python
newdf.groupby(['KV'])[['T5', 'T6']].agg(['min', 'mean'])
```

```python
newdf.groupby(['KV'])[['T5', 'T6']].agg(['min', 'max'])
```

```python
# Tổng hợp trên tất cả các biến số có thể thao tác 1 hàm tổng hợp của gom nhóm
df_gr = newdf.groupby(['KV'], as_index=False).count()
df_gr
```

```python
pd.pivot_table(newdf, values=['T5', 'T6'], columns='KV', aggfunc=['min', 'mean', 'max'])
```

```python
pd.pivot_table(newdf, values=['T5', 'T6'], columns=['KV', 'KT'], aggfunc=['min', 'mean', 'max'])
```
